```python
import matplotlib.pyplot as plt
import numpy as np
```

### Accuracy VS Dimension Multiplier for One Layer OHN

```python
multipliers = np.array([12,20,40,80,160])
w_threshold_acc = np.array([96.8, 97.3, 97.5, 97.6, 97.8])
w_o_threshold_acc = np.array([89,88,88,87,88])
```

```python
plt.plot(multipliers, w_threshold_acc, 'o-', color='lightgreen',label = "With Threshold")
plt.plot(multipliers, w_o_threshold_acc, 'o-', color='pink', label = "Without Threshold")
plt.legend()
plt.title("Accuracy VS Dimension Multiplier for One Layer OHN")
plt.xlabel("Dimension Multiplier")
plt.ylabel("% Accuracy")
plt.show()
```

### Accuracy VS Gaussian Noise Level on Weights

```python
noises = np.array([0, 0.02, 0.05, 0.1, 0.2, 0.5])
OHN20 = np.array([97.3, 95.7, 94.4, 93.4, 91.8, 89.2])
OHN80 = np.array([97.6, 96.2, 96.1, 94.5, 93.3, 90.4])
MLP784 = np.array([97.6, 93.3, 88.4, 82.1, 77.2, 73.3])
MLP72 = np.array([97.2, 84.3, 62.5, 21.1, 10.7, 10.5])
MLP20t784 = np.array([97.7, 95.7, 94.9, 94.3, 93.9, 93.2])
```

```python
plt.plot(noises, OHN20, 'o-', color='lightgreen',label = "OHN-20")
plt.plot(noises, OHN80, 'o-', color='pink', label = "OHN-80")
plt.plot(noises, MLP784, 'o-', color='bisque',label = "1L-MLP(784->784->10)")
plt.plot(noises, MLP72, 'o-', color='firebrick', label = "1L-MLP(784->72->10)")
plt.plot(noises, MLP20t784, 'o-', color='silver',label = "1L-MLP(784->20*784->10)")
plt.legend()
plt.title("Test Accuracy VS Gaussian Noise Level on Weights")
plt.xlabel("Noise level (standard deviation)")
plt.ylabel("% Test Accuracy")
plt.show()
```
